```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.datasets import load_files
```

```python
with open('stop_word_list.txt', 'r') as f:
    stop_word_list = f.read().split('\n')
print(stop_word_list)
```

```python
services=load_files(container_path="services")
service_list=services.filenames
vectorizer =  TfidfVectorizer(stop_words=stop_word_list)
tfidf_data = vectorizer.fit_transform(services.data)
feature_names = vectorizer.get_feature_names()
print(tfidf_data.shape)
```

```python
def fearture_extraction(file_index):
    print(service_list[file_index])
    print("\n")
    for col in tfidf_data.nonzero()[1]:
        if tfidf_data[file_index,col]>0.1:
            print (feature_names[col], ' - ', tfidf_data[file_index,col])
    print("-------------------------------------------")
```

```python
for i in range(10):
    fearture_extraction(i)
```

```python
def fearture_extractor_txt(file_index):
    filename= service_list[file_index][:-4] +"_result.txt"
    with open(filename, 'w') as f:
        f.write(service_list[file_index])
        f.write("\n")
        for col in tfidf_data.nonzero()[1]:
            if tfidf_data[file_index,col]>0.1:
                f.write (feature_names[col])
                f.write(' - ')
                f.write(str(tfidf_data[file_index,col]))
                f.write("\n")
```

```python
for i in range(10):
    fearture_extractor_txt(i)
```

```python
import csv
def fearture_extractor_csv(file_index):
    filename= service_list[file_index] +".csv"
    with open(filename, 'w') as f:
        writer = csv.writer(f, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
        for col in tfidf_data.nonzero()[1]:
            if tfidf_data[file_index,col]>0.1:
                writer.writerow([feature_names[col],tfidf_data[file_index,col]])
```

```python
for i in range(10):
    fearture_extractor_csv(i)
```
