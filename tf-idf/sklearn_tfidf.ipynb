```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.datasets import load_files
```

```python
with open('stop_word_list.txt', 'r') as f:
    stop_word_list = f.read().split('\n')
print(stop_word_list)
```

```python
services=load_files(container_path="services")
service_list=services.filenames
vectorizer =  TfidfVectorizer(stop_words=stop_word_list)
tfidf_data = vectorizer.fit_transform(services.data)
feature_names = vectorizer.get_feature_names()
print(tfidf_data.shape)
```

```python
def fearture_extraction(file_index):
    print(service_list[file_index])
    print("\n")
    for col in tfidf_data.nonzero()[1]:
        if tfidf_data[file_index,col]>0.1:
            print (feature_names[col], ' - ', tfidf_data[file_index,col])
    print("-------------------------------------------")
```

```python
for i in range(5):
    fearture_extraction(i)
```
